<h2> Use the attribute information provided in 
auto-mpg.names to name your pandas 
DataFrame columns. </h2>

```python
import pandas as pd

df = pd.DataFrame(X)
#The column headers are named per below.
df
```

<h2>Some of the horsepower data is marked as 
missing with a ‘?’. You need to replace this ‘?’ 
with a pandas-appropriate missing value, then 
convert the column to numeric.</h2>

```python
#Find the number of missing values in horsepower column.
df["horsepower"].isnull().sum()
```

```python
df["horsepower"]=df["horsepower"].replace(["nan"],[0] )
df
```

<h2>. The origin column values are 1, 2, and 3. 
Convert these values respectively to ‘USA’, 
‘Asia’, and ‘Europe’</h2>

```python
#Replace origin values 1,2,3 with USA, ASIA, EUROPE
df["origin"]=df["origin"].replace([1, 2, 3], ['USA', 'ASIA', 'EUROPE'])
df
```

<h2> Create a scatterplot that shows the 
relationship between horsepower and weight.</h2>

```python
# Import seaborn
import seaborn as sns

# Create a visualization of Cylinders per model year and origin
sns.relplot(data=df, x="model_year", y="origin", hue="cylinders", kind="line")
```

<h2> Create a scatterplot that shows the 
relationship between horsepower and weight.</h2>

```python
sns.lmplot(x='horsepower', y='weight', data=df, fit_reg=True)
```
