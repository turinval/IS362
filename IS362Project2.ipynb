<h3> For the 2015 file two unique cities with separate price and quantity columns must be combined.</h3>

This is achieved by first renaming first 3 columns. Add a column city with first city name.
Then creating two slices, one with Bloomington and another with Muncie.
This time the city name has to be added in the slice itself. So the slice with just Bloomington data should say city Bloomington. Similarly for Muncie.
Then merge the two city in one data frame.
FInally merge to the original data frame.

```python
c2015
#AS before, rename columns clear incorrect row with cloumn names and empty rows. But just for first 4 columns.
c2015.rename(columns={"Unnamed: 0":"month","Unnamed: 1":"day","Noblesville":"price","Unnamed: 3":"quantity" }, inplace=True)


#Clean up row with column names and blank row.Note city column is created here in assign function.
n02015= c2015.drop([0,54]).assign(city="Noblesville")

#Extract all the Bloomington Price and Quantity columns and rename. This time include city column. 
#This will enable assign function to assign Bloomington city to the slice.
Bloomington= n02015.loc[:,['month','day','Bloomington', 'Unnamed: 5','city']].rename(columns={"Bloomington":"price", "Unnamed: 5": "quantity"})
Bloomington1=Bloomington.assign(city="Bloomington")

#Extract all Muncie Price and Quantity columns and rename with city column as well.
#Assign city name Muncie to this slice.
Muncie=n02015.loc[:,['month','day','Muncie', 'Unnamed: 7','city']].rename(columns={"Muncie":"price", "Unnamed: 7": "quantity"})
Muncie1=Muncie.assign(city="Muncie")

#Use the merge function to combine Muncie slice and Bloomington slice with outer join. Only outer join will work.
#Advantage of Cartesian product(of outer join) is seperate columns with same column names will be concatenated.
BloomMuncie=Bloomington1.merge(Muncie1, how='outer')

#Finally merge the combined Bloomington and Muncie (Price and Quantity data) with the Noblesville price and quantity data frame.
#Must use outer join again.
n2015=n02015.merge(BloomMuncie, how='outer')

#Remove redundant columns.
n2015= n2015.drop(columns=['Bloomington', 'Unnamed: 5', 'Muncie', 'Unnamed: 7'])
#Below shows all three cities were combined successfully based on month, day, price, and quantity columns.
n2015.groupby(['city']).count()
```

<h1> The CSV files for 2010 to 2015 were tidied up. Now combine all the files together with another merge</h1>
<h2> The goal was to show the manual tidying process although a function could be made in a for loop with the column and row index by finding the shape() function and then iterating through the renaming and merge for each city.</h2>

```python
#add the corresponding year to each tidied csv file and call the column 'year'. 
#Relabel as new dataframe using y infront so the year column is retained. This keeps the correct year for each tidied dataframe.
y2010=n2010.assign(year=2010)
y2011=n2011.assign(year=2011)
y2012=n2012.assign(year=2012)
y2013=n2013.assign(year=2013)
y2014=n2014.assign(year=2014)
y2015=n2015.assign(year=2015)

#concatenate the new dataframe with years together.
df2010df2015= pd.concat([y2010, y2011,y2012,y2013,y2014,y2015])

#Get basic info of combined dataset from 2010 to 2015.
df2010df2015.describe()

```
