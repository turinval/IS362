<h3> Study Vulnerabilities associated to Ransomware. </h3>
 
 Strengths of the CISA known exploits table are the ability to capture semantic data quickly for risks and responses.
 Weaknesses is the CISA KEV CSV file is not a detailed statistical study of vulnerabilities.
 Opportunities is the ability to use both NIST data and CISA data for cross comparison.
 Threats is the lack of required actions to prevent against ransomware and direct access to virtual and physical data. 
 
 Below we will conclude the CISA KEV to visualize the Ransomware trait in the table.

```python
ransom1= df.loc[df['knownRansomwareCampaignUse']=="Known" ]
ransom2=ransom1.loc[ransom1['requiredAction'] != "Apply updates per vendor instructions."]
ransom3=ransom2.loc[ransom1['requiredAction'] != "The impacted product is end-of-life and should be disconnected if still in use."]
ransom3
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns 

LE = LabelEncoder()
ransom3['graphxcode'] = LE.fit_transform(ransom3['requiredAction'])
ransom3['graphycode'] = LE.fit_transform(ransom3['product'])
ransom3['graphx1code'] = LE.fit_transform(ransom3['knownRansomwareCampaignUse'])
sns.catplot(data=ransom3, x="graphxcode", y="graphycode", hue="dateAdded", kind="box")
```

<h3> Vulnerabilities with a known ransomware attack with only requiredAction as "apply mitigations</h3>

You will see "Apply mitigations per vendor" as number 2 in graphxcode and Products are listed in graphycode.
So regardless of the product the mitigations are listed as the only solution. This is a result of the filters on the table.
But this is important to note. Because these mitigations are important to reduce the risk of cybersecurity threats of ransomware.

We will conclude by usiing a SciKit learn Decision tree.

```python
from sklearn import tree
ransom4=ransom3.query('dateAdded >= "2022-01-01" and dateAdded <= "2024-05-18"')
X=ransom4[['graphxcode','graphx1code']]
Y=ransom4['graphycode']
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
clf.predict_proba([[2., 2.]])
tree.plot_tree(clf)
```

<h3> Conclusion: Mitigation of vulnerabilities is the most predicted Required Action </h3>

The graph above mirrors the decision tree below. Remember the data is filtered to show only known ransomware attacks.
Mitigation of of vulnerabilities with date added after 1/1/2024 is another filter applied to reduce the data from the output.

Compare this with the decision tree of the original data with encoded data. The decision tree is all the way after the text in 446 below. Notice the difference in complexity of tree. With the filtered data from the above tree, its much easier to see the importance of mitigation in one graph than with unfiltered data.

```python
#Transform original df data frame into column categories and turn into codes for scikit learn.
df['product'] = pd.Categorical(df['product'])
df['product'] = df['product'].cat.codes
df['vulnerabilityName'] = pd.Categorical(df['vulnerabilityName'])
df['vulnerabilityName']=df['vulnerabilityName'].cat.codes
df['requiredAction'] = pd.Categorical(df['requiredAction'])
df['requiredAction']=df['requiredAction'].cat.codes
df['knownRansomwareCampaignUse'] = pd.Categorical(df['knownRansomwareCampaignUse'])
df['knownRansomwareCampaignUse']=df['knownRansomwareCampaignUse'].cat.codes
df
```

```python
from sklearn import tree
X=df[['requiredAction','product']]
Y= df['knownRansomwareCampaignUse']
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
clf.predict_proba([[2., 2.]])
tree.plot_tree(clf)
```
