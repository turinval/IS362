<h1> What is the Easternmost airport in the US</h1> 

<h3> It is debatable what is meant easternmost airport in United States in reference to what?</h3> 

While North has the north pole and south the south pole, east and west depends on what latitude and longitude is defined as the center of reference. Since the question states center of the United States, the assumption is the center of the United States is the referenced base map, and not the global map. 


```python
#The center of the United States map is 39.82844474413941, -98.57855702371702 (Per Google Maps.)
#Eastward of this point is going to the East Coast. This can be found in tz =-5.  
df.loc[((df['lon'] <= -50) & (df['lon'] > -69)&(df['tz']==-5))]
```

<h1> Eastport Municipal Airport is the Easternmost airport in the US.</h1>

As before just using the nlargest won't work. And exploring the ranges of latitude and longitude to find the East coast of the United States using the lat 39 and lon-98 as the center point of reference. (Assuming only the US map is being looked at, not worldwide.)

We find the East coast of the US in terms of ranges for latitude and longitude. This can be easily found using tzone as America/New York. Then find the largest longitude in this subset to find the easternmost airport.   

```python
df.loc[((df['lon'] <= -50) & (df['lon'] > -69)&(df['tz']==-5))].nlargest(1,'lon')
```

<h1> On February 12th, 2013, which New York area airport had the windiest weather?</h1>

```python
#Load the weather CSV file.
weather = r"C:\Users\turin\Downloads\weather.csv"
wdf= pd.read_csv(weather)
#Explore the connection of weather and airports dataframes.
#print(wdf.head(),df.head())
wdf.groupby(['origin']).count()
```

```python
#The only matching column is origin from Weather table and faa from airports table.
#Since weather.csv only has NYC airports, filter airports to show only these airports.
nyairports= df.query('faa == ["EWR", "JFK", "LGA"]') 
#use merge to combine the airports and weather table
wdf.rename(columns = {'origin ':'faa'}, inplace = True)
#This took a long time and lot of adjustments to get working.
weatherairports= wdf.merge(nyairports, left_on='origin',right_on='faa')
weatherairports.groupby(['origin']).count()
```

```python
#Find the weather at the three airports (EWR, JFK, LGA) on the date of February 12, 2013.
windynyc=weatherairports.loc[(weatherairports['month']==2) & (weatherairports['day']==12) & (weatherairports['year']==2013)]
windynyc
```

<h1> The windiest airport on February 12, 2013 in NY was Newark Liberty Airport</h1>

```python
windynyc.nlargest(1,'wind_speed') 
```
