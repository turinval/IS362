<h2> Your task is to create a pandas DataFrame that displays Customers’ Last 
Name and First Name, and each customer’s purchased Track names and 
Album Titles. The information should be sorted by Customer LastName 
then Customer FirstName. </h2>

```python
fdf= pd.read_sql("SELECT LastName, FirstName, Name, Title FROM customers INNER JOIN invoices ON Invoices.CustomerId= customers.CustomerId INNER JOIN invoice_items ON invoices.InvoiceId= invoice_items.InvoiceId INNER JOIN tracks ON tracks.TrackId = invoice_items.TrackId INNER JOIN albums ON albums.AlbumId =tracks.AlbumId ORDER BY  LastName, FirstName", cnk)
fdf
```

<h1> See every customers favorite movie </h1>

```python
fdf.groupby(['LastName']).describe()
#Appears in the top column when using describe() function.
```

<h1> Run a query for unit prices and quantity per customer </h1>

```python
cdf= pd.read_sql("SELECT InvoiceDate,  Name, Title, Quantity, tracks.UnitPrice FROM customers INNER JOIN invoices ON Invoices.CustomerId= customers.CustomerId INNER JOIN invoice_items ON invoices.InvoiceId= invoice_items.InvoiceId INNER JOIN tracks ON tracks.TrackId = invoice_items.TrackId INNER JOIN albums ON albums.AlbumId =tracks.AlbumId ORDER BY  LastName, FirstName", cnk)
ccdf= cdf.groupby(["InvoiceDate"]).sum()
ccdf
```

<h1>Visualize in seaborn</h1>

```python
import seaborn as sns
sns.set_theme()
sns.scatterplot(data=ccdf, x="Quantity", y="UnitPrice")


#An ideal unit price to Quantity ratio would help define meaningful sales.
```

```python
sns.catplot(data=ccdf, x="InvoiceDate", y="UnitPrice")

#UNit Price by Invoice Date. Knowing what dates the frequency of purchases are consistent for each unit price.
```
